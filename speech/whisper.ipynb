패키지 불러오기

```python
from IPython.display import Audio

import whisper
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"[INFO] Using device: {device}")
```

사전 학습 모델 불러오기

```python
# model = whisper.load_model("large")

# https://github.com/openai/whisper?utm_source=turingpost.co.kr&utm_medium=referral&utm_campaign=topic-15-ai-whisper 참고
# Size	Parameters	English-only model	Multilingual model	Required VRAM	Relative speed
# tiny	39 M	tiny.en	tiny	~1 GB	~10x
# base	74 M	base.en	base	~1 GB	~7x
# small	244 M	small.en	small	~2 GB	~4x
# medium	769 M	medium.en	medium	~5 GB	~2x
# large	1550 M	N/A	large	~10 GB	1x
# turbo	809 M	N/A	turbo	~6 GB	~8x

model = whisper.load_model("c").to(device)
```

예제 음성 데이터 확인

```python
audio_path = "examples/example.wav"
Audio(audio_path)
```

위스퍼 모델 추론

```python
# Whisper 모델은 ffmpeg를 사용하여 오디오 파일을 처리합니다.
# 먼저 ffmpeg 이 시스템에 없다면 설치
# sudo apt install ffmpeg

result = model.transcribe(audio_path)
result["text"]
```
